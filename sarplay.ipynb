```python
import numpy as np
import matplotlib.pyplot as plt
import struct

def read_cphd(filename, num_samples=1024, num_lines=1024):
    """
    Reads a CPHD file assuming complex64 (float32 pairs) encoding.
    Adjust num_samples and num_lines based on your data dimensions.
    """
    with open(filename, 'rb') as f:
        data = np.fromfile(f, dtype=np.complex64, count=num_samples * num_lines)
        data = data.reshape((num_lines, num_samples))
    return data

def visualize_cphd(data):
    """Visualizes the magnitude and phase of CPHD SAR data."""
    magnitude = np.abs(data)
    phase = np.angle(data)
    
    fig, axes = plt.subplots(1, 2, figsize=(12, 6))
    
    axes[0].imshow(20 * np.log10(magnitude + 1e-6), cmap='gray')
    axes[0].set_title('Magnitude (dB)')
    
    axes[1].imshow(phase, cmap='jet')
    axes[1].set_title('Phase')
    
    plt.show()

```

```python
filename = "2023-08-31-01-09-38_UMBRA-04_CPHD.cphd"
cphd_data = read_cphd(filename)
visualize_cphd(cphd_data)
```
